```python
#14
a = 3
b = 4

a *= 2
b -= a

print(a > b and b > 0 )
# solution

#Step-by-step:
a *= 2   # a = a * 2 → 6
b -= a   # b = b - a → 4 - 6 = -2
a = 3, b = 4

a *= 2 → a = 6

b -= a → b = -2

Condition:

a > b → 6 > -2 → True

b > 0 → -2 > 0 → False

True and False → False
```

```python
# MATCH CASE 
num = 4

match num :
    case 1 :
        print("one")
    case 2 :
        print("two")
    case 3 :
        print("three")
    case 4 :
        print("four")
    case 5 :
        print("five")
    case _ :
        print("enter value from 1 to 5")
```
